```python
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister
```

```python
qc = QuantumCircuit(3)
```

```python
alice_register = QuantumRegister(2,name="alice")
bob_register = QuantumRegister(1, name="bob")
qc = QuantumCircuit(alice_register, bob_register)
```

```python
qc.h(alice_register[1])
qc.cx(alice_register[1],bob_register[0])
```

```python
aclassical_register = ClassicalRegister(1)
qc.add_register(aclassical_register)

```

```python
qc.cx(alice_register[0],alice_register[1])
qc.h(alice_register[0])
qc.draw('mpl')
```

```python
qc.measure(alice_register[0],aclassical_register[0])
qc.measure(alice_register[1],aclassical_register[0])
qc.draw('mpl')
```

```python
with qc.if_test((aclassical_register,0b01)):
        pass
with qc.if_test((aclassical_register,0b01)):
        qc.x(bob_register[0])
with qc.if_test((aclassical_register,0b10)):
        qc.z(bob_register[0])
with qc.if_test((aclassical_register,0b11)):
        qc.x(bob_register[0])
        qc.z(bob_register[0])
```

```python
qc.draw('mpl')
```
